```python
import pandas as pd
df = pd.read_csv("gs://project_major/Major_Project/DATA.csv")
```

```python
df = df.iloc[:, 1:]
```

```python
df
```

```python
pd.set_option('display.max_columns', None)
```

```python
df['ID'] = df.index
```

```python
df
```

```python
df.columns
```

```python
datetime_dim = df[['incident_datetime']].reset_index(drop = True)
```

```python
datetime_dim
```

```python
datetime_dim['incident_datetime'] = pd.to_datetime(df['incident_datetime'])
```

```python
datetime_dim['day'] = datetime_dim['incident_datetime'].dt.day
datetime_dim['month'] = datetime_dim['incident_datetime'].dt.month
datetime_dim['year'] = datetime_dim['incident_datetime'].dt.year
datetime_dim['quarter'] = datetime_dim['incident_datetime'].dt.quarter
datetime_dim['weekday'] = datetime_dim['day'].apply(lambda x: 'y' if x < 5 else 'n')
datetime_dim['weekend'] = datetime_dim['day'].apply(lambda x: 'n' if x < 5 else 'y')


datetime_dim['datetime_id'] = datetime_dim.index
```

```python
datetime_dim
```

```python
datetime_dim.columns
```

```python
time_dim = df[['incident_datetime']].reset_index(drop = True)
time_dim['incident_datetime'] = pd.to_datetime(df['incident_datetime'])
time_dim['time'] = time_dim['incident_datetime'].dt.strftime('%H:%M:%S')
time_dim['hour'] = time_dim['incident_datetime'].dt.hour
time_dim['minute'] = time_dim['incident_datetime'].dt.minute
time_dim['second'] = time_dim['incident_datetime'].dt.second
time_dim['time_id'] = time_dim.index
```

```python
time_dim
```

```python
time_dim.columns
```

```python
location_dim = df[['latitude', 'longitude']].reset_index(drop=True)
location_dim['location_id'] = location_dim.index
```

```python
location_dim
```

```python
location_dim.columns
```

```python
incident_dim = df[['incident_id', 'incident_code', 'incident_category', 'incident_subcategory', 'incident_description']].reset_index(drop=True)
incident_dim['Inci_id'] = incident_dim.index
```

```python
incident_dim
```

```python
incident_dim.columns
```

```python
resolution_dim = df[['resolution']].reset_index(drop=True)
resolution_dim['resolution_id'] = location_dim.index
```

```python
resolution_dim
```

```python
resolution_dim.columns
```

```python
fact_table = df.merge(datetime_dim, left_on='ID', right_on='datetime_id') \
             .merge(time_dim, left_on='ID', right_on='time_id') \
             .merge(location_dim, left_on='ID', right_on='location_id') \
             .merge(incident_dim, left_on='ID', right_on='Inci_id') \
             .merge(resolution_dim, left_on='ID', right_on='resolution_id')\
```

```python
fact_table
```

```python
fact_table.columns
```

```python
complete_fact_table = fact_table[['incident_datetime_x', 'incident_date', 'incident_time',
       'incident_id_x', 'incident_number', 'report_type_code', 'incident_code_x', 'incident_category_x',
       'incident_subcategory_x', 'incident_description_x', 'resolution_x', 'police_district',
       'latitude_x', 'longitude_x', 'ID', 'day', 'month', 'year', 'quarter', 'weekday', 'weekend', 'datetime_id',
       'incident_datetime', 'time', 'hour', 'minute', 'second', 'time_id',
       'latitude_y', 'longitude_y', 'location_id', 'Inci_id', 'resolution_id']]
```
